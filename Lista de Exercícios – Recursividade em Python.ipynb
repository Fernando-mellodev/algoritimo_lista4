```python
"""1. Escreva uma função recursiva que calcule a soma dos dígitos de um número inteiro positivo.
Exemplo: 123 =>6"""

def soma_digitos (n):
    if n < 10:
        return n
    return (n % 10) + soma_digitos(n // 10)
print (f"soma dos digitos de 123: {soma_digitos(123)}")
```

```python
"""2. Crie uma função recursiva que conte quantos dígitos tem um número inteiro.
Exemplo: 9876 =>4"""

def contar_digitos(n):
    if n < 10:
        return 1
    return 1 + contar_digitos(n // 10)

print (f"Digitos em 9878: {contar_digitos(9876)}")
print (f"Digitos em 5: {contar_digitos(5)}")
```

```python
"""5. Crie uma função recursiva que calcule o produto de dois números usando apenas soma.
Exemplo: 3 × 4 =>12"""

def produto_soma(a, b):
    if b == 0:
        return 0
    return a + produto_soma(a, b -1)
print(f"Produto de 3 x 4: {produto_soma(3, 4)}")
print(f"Produto de 5 x 2: {produto_soma(5, 2)}")
```

```python
"""6. Implemente uma função recursiva que encontre o maior elemento de uma lista.
"""

def maior_elemento(lista):
    if len(lista) == 1:
        return lista[0]
    maior_resto = maior_elemento(lista[1:])

    if lista[0] > maior_resto:
        return lista[0]
    return maior_resto

print(f"Maior de [1, 5, 3]: {maior_elemento([1, 5, 3])}") 
print(f"Maior de [10, 2, 45, 8]: {maior_elemento([10, 2, 45, 8])}")
```

```python
"""7. Escreva uma função recursiva que calcule o máximo divisor comum (MDC) entre dois
números (algoritmo de Euclides)."""

def mdc(a, b):
    if b == 0:
        return a
    return mdc(b, a% b)

print(f"MDC de 48 e 18: {mdc(48, 18)}")
print(f"MDC de 20 e 5: {mdc(20, 5)}")
```

```python
"""8. Crie uma função recursiva que conte quantas vezes um elemento aparece em uma lista.
"""
def contar_elemento(lista, item):
    if not lista:
        return 0

    encontrou = 1 if lista[0] == item else 0
    return encontrou + contar_elemento(lista[1:], item)

minha_lista = [1, 2, 1, 3, 1]
print(f"O número 1 aparece {contar_elemento(minha_lista, 1)} vezes.") 
print(f"O número 5 aparece {contar_elemento(minha_lista, 5)} vezes.")
```

```python
"""9. Implemente uma função recursiva que gere a soma de todos os elementos de uma lista.
"""
def soma_lista(lista):
    if not lista:
        return 0
    
    return lista[0] + soma_lista(lista[1:])

numeros = [1,2,3,4,5]

print(f"A soma de {numeros} é: {soma_lista(numeros)}") 
print(f"A soma de [10, 20] é: {soma_lista([10, 20])}")
```

```python
"""10. Escreva uma função recursiva que determine se um número é primo.
"""
def eh_primo(n, divisor=2):
    if n <= 1:
        return False
    if divisor * divisor > n:
        return True
    if n % divisor ==0:
        return False
    return eh_primo(n, divisor +1)


print(f"7 é primo? {eh_primo(7)}")   
print(f"10 é primo? {eh_primo(10)}") 
print(f"13 é primo? {eh_primo(13)}")
```

```python
"""11. Soma de números em um intervaloCrie uma função que receba dois números, a e b, e retorne a soma de todos os números inteiros entre eles (inclusive).Exemplo: soma_intervalo(3, 6) => $3 + 4 + 5 + 6 = 18$"""


def soma_intervalo (a, b):
    if a > b:
        return 0
    return a + soma_intervalo(a +1,b)

print (f'soma de 3 a 6:{soma_intervalo(3,6)}')
```

```python
"""12. Potência RecursivaImplemente uma função que calcule $a^b$ (a elevado a b) usando apenas multiplicações sucessivas.
Exemplo: potencia(2, 3) => $2 \times 2 \times 2 = 8$"""

def potencia(base, exp):
    if exp ==0:
        return 1
    return base * potencia(base, exp -1)

print (f'2 elevado a 3: {potencia(2,3)}')

```

```python
"""13. Contar letras maiúsculas em uma string
Dada uma string, conte quantas letras maiúsculas ela possui.

Exemplo: contar_maiusculas("PyThon") => 2"""

def contar_maiuscula(s):
    if not s:
        return 0
    contagem = 1 if s[0].isupper() else 0
    return contagem + contar_maiuscula(s[1:])

print (f"Maiúsculas em 'Software Engineering': {contar_maiuscula('Software Engineering')}")
```

```python
"""14. Converter Decimal para Binário
Crie uma função que receba um número inteiro e retorne sua representação binária como string.
Exemplo: decimal_para_binario(10) => "1010"""

def decimal_binario(n):
    if n == 0:
        return ""
    return decimal_binario(n//2) + str(n%2)

print (f"2656 em binario: {decimal_binario(2656)}")
```

```python
"""15. Verificar se uma lista está ordenada
Crie uma função que retorne True se a lista estiver em 
ordem crescente e False caso contrário.
Exemplo: esta_ordenada([1, 2, 8, 4]) => False"""

def ordenada(lista):
    if len(lista) <= 1:
        return True
    if lista[0] > lista[1]:
        return False
    return ordenada(lista[1:])

print(f"[1, 2, 5] ordenada? {ordenada([1, 2, 5])}") 
print(f"[1, 5, 2] ordenada? {ordenada([1, 5, 2])}")
```
